# Encoding

```python
fraud['Month'] = fraud['Month'].replace({
    'Jan' : 0,
    'Feb' : 1,
    'Mar' :2,
    'Apr' : 3,
    'May': 4, 
    'Jun': 5,
    'Jul': 6,
    'Aug' : 7,
    'Sep': 8,
    'Oct' : 9,
    'Nov' : 10,
    'Dec' : 11
})

fraud['DayOfWeek'] = fraud['DayOfWeek'].replace({
    'Monday' : 0,
    'Tuesday': 1,
    'Wednesday': 2,
    'Thursday': 3,
    'Friday': 4,
    'Saturday': 5,
    'Sunday':6
})

fraud['Make'] = fraud[['Make']].replace({
    'Lexus': 0,
    'Ferrari': 1, 
    'Mecedes': 2,
    'Porche': 3,
    'Jaguar': 4,
    'BMW': 5,            
    'Nisson': 6,
    'Saturn': 7,
    'Mercury':8,
    'Dodge' : 9,
    'Saab' : 10,
    'VW' : 11,
    'Ford': 12,
    'Accura': 13,
    'Chevrolet': 14,
    'Mazda' : 15,
    'Honda' : 16,
    'Toyota' : 17,
    'Pontiac': 18
    })   

fraud['AccidentArea'] = fraud['AccidentArea'].replace( {
    'Rural' : 0,
    'Urban' : 1
    })


fraud['DayOfWeekClaimed'] = fraud['DayOfWeekClaimed'].replace( {
    'Monday': 0,
    'Tuesday': 1,
    'Wednesday': 2,
    'Thursday': 3,
    'Friday': 4, 
    'Saturday': 5,
    'Sunday' : 6
    })

fraud['MonthClaimed'] = fraud['MonthClaimed'].replace( {
    'Jan' : 0,
    'Feb' : 1,
    'Mar' :2,
    'Apr' : 3,
    'May': 4, 
    'Jun': 5,
    'Jul': 6,
    'Aug' : 7,
    'Sep': 8,
    'Oct' : 9,
    'Nov' : 10,
    'Dec' : 11
    })

fraud['Sex'] = fraud['Sex'].replace( {
    'Female' : 0,
    'Male' : 1
    })
    

fraud['MaritalStatus'] = fraud['MaritalStatus'].replace( {
    'Widow' : 0,
    'Divorced' : 1,
    'Single' : 2,    
    'Married' : 3
    })


fraud['Fault'] = fraud['Fault'].replace( {
    'Third Party' : 0,
    'Policy Holder' : 1
    })

fraud['PolicyType'] = fraud['PolicyType'].replace( {
    'Sport - Liability' : 0,
    'Sport - All Perils' : 1,
    'Utility - Liability' : 2,
    'Utility - Collision' :3,
    'Utility - All Perils' :4,
    'Sport - Collision' : 5,
    'Sedan - All Perils' : 6 ,
    'Sedan - Liability' : 7,
    'Sedan - Collision' : 8
    })


fraud['VehicleCategory'] = fraud['VehicleCategory'].replace( {
    'Utility' : 0,
    'Sport' : 1, 
    'Sedan' : 2
    })

fraud['VehiclePrice'] = fraud['VehiclePrice'].replace( {
    'less than 20000' : 0,
    '20000 to 29000': 1,
    '30000 to 39000': 2,
    '40000 to 59000': 3,
    '60000 to 69000' : 4,
    'more than 69000' : 5
    })  
  

fraud['Days_Policy_Accident'] = fraud['Days_Policy_Accident'].replace( {
    'none' : 0,
    '1 to 7' : 1,
    '8 to 15' : 2,
    '15 to 30' :3,
    'more than 30' :4
    })

fraud['Days_Policy_Claim'] = fraud['Days_Policy_Claim'].replace( {
    'none' : 0,
    '8 to 15' : 1,
    '15 to 30' : 2,
    'more than 30' : 3
    })


fraud['PastNumberOfClaims'] = fraud['PastNumberOfClaims'].replace( {
    'none': 0,
    '1': 1,
    '2 to 4': 2,
    'more than 4': 3
    })

fraud['AgeOfVehicle'] = fraud['AgeOfVehicle'].replace({
    'new' : 0,
    '2 years' : 1,
    '3 years' : 2,
    '4 years' : 3,
    '5 years' : 4, 
    '6 years' : 5,
    '7 years' : 6,      
    'more than 7' : 7
    })

fraud['AgeOfPolicyHolder'] = fraud['AgeOfPolicyHolder'].replace({
    '16 to 17' : 0,
    '18 to 20' : 1,  
    '21 to 25' : 2,
    '26 to 30' : 3,
    '31 to 35' : 4,
    '36 to 40' : 5,
    '41 to 50' : 6,
    '51 to 65' : 7, 
    'over 65' : 8
    })

fraud['PoliceReportFiled'] = fraud['PoliceReportFiled'].replace({
    'Yes' : 0,
    'No' : 1
    })


fraud['WitnessPresent'] = fraud['WitnessPresent'].replace({
    'Yes' : 0,
    'No' : 1
    })


fraud['AgentType'] = fraud['AgentType'].replace({
    'Internal' : 0,
    'External' : 1
    })


fraud['NumberOfSuppliments'] = fraud['NumberOfSuppliments'].replace({
    'none' : 0,
    '1 to 2' : 1,
    '3 to 5' : 2,
    'more than 5' : 3
    })


fraud['AddressChange_Claim'] = fraud['AddressChange_Claim'].replace({
    'no change' : 0,
    'under 6 months' : 1,
    '1 year' : 2,
    '2 to 3 years' : 3,
    '4 to 8 years' : 4
    })

fraud['NumberOfCars'] = fraud['NumberOfCars'].replace({
    '1 vehicle' : 0,
    '2 vehicles' : 1,
    '3 to 4' : 2,
    '5 to 8' : 3,
    'more than 8' : 4
    })


fraud['BasePolicy'] = fraud['BasePolicy'].replace({
    'All Perils' : 0, 
    'Liability' : 1,
    'Collision' : 2 
    })
```

# Stratified K_Fold cross validation

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

# Stratified K_Fold cross validation on logistic regression

X = fraud.drop('FraudFound_P', axis = 1)
y = fraud['FraudFound_P']

skf = StratifiedKFold(n_splits = 5, shuffle=True, random_state=42)
log_model = LogisticRegression()

for train_index, test_index in skf.split(X, y):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    
    
    sc = StandardScaler()
    X_train = sc.fit_transform(X_train)
    X_test = sc.transform(X_test)
    
    log_model.fit(X_train, y_train)
    y_predict = log_model.predict(X_test)
    print('score :', log_model.score(X_test, y_test))
    print('\n')
    print('confusion matrix :\n', confusion_matrix(y_test, y_predict))
    print('\n')
    print('classification_report :\n', classification_report(y_test, y_predict))
    print('\n')
    print('accuracy score :\n', accuracy_score(y_test, y_predict))
    print('\n')
    print('precision score:\n', precision_score(y_test, y_predict))
    print('\n')
    print('recall score :\n', recall_score(y_test, y_predict))
    print('\n')
    print(' f1 score :\n', f1_score(y_test, y_predict))
    print('\n')
    print('\n')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn.utils import resample
```

```python
class BaggingClassifier:
    def __init__(self, base_estimator=DecisionTreeClassifier(), n_estimators=10, max_samples=1.0):
        self.base_estimator = base_estimator
        self.n_estimators = n_estimators
        self.max_samples = max_samples
        self.estimators_ = []

    def fit(self, X, y):
        for i in range(self.estimators):
            X_subset, y_subset = resample(X, y, n_samples=int(self.max_samples * X.shape[0]), replace=True)
            estimator = self.base_estimator.fit(X_subset, y_subset)
            self.estimators_.append(estimator)

    def predict(self, X):
        predictions = []
        for estimator in self.estimators_:
            predictions.append(estimator.predict(X))
        return np.mean(predictions, axis=0).round().astype(int)
```
